```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("F:/Datasets/homeprices.csv")
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df,dummies],axis='columns')
merged
```

```python
final = merged.drop(['town','west windsor'],axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = final.drop('price',axis='columns')
X
```

```python
y = final.price
y
```

```python
model.fit(X,y)
```

```python
model.predict([[2800,0,1]]) #robbinscille
```

```python
model.predict([[3400,0,0]]) #west windsor
```

```python
model_score = model.score(X,y) * 100
model_score
```
